```python
import cv2
import matplotlib.pyplot as plt
from src import walls_detection_yolo as wdy
```

```python
files = ["../Screenshot 2024-09-26 at 01.37.20.png", 
         "data/images/test/colorful_1334_F3_original.png"]

def plot_images(image1, image2, text1="", text2=""):

    plt.figure(figsize=(15, 15))

    plt.subplot(1, 2, 1)
    plt.title(text1)
    plt.imshow(image1)

    plt.subplot(1, 2, 2)
    plt.title(text2)
    plt.imshow(image2)

    plt.show()

for file in files:
    image = cv2.imread(file)
    image_done = wdy.get_walls(image.copy())

    plot_images(image, image_done, "image original", "image_segmented")
```
