```python
from glob import glob
import pandas as pd
import os

files = glob("dataset/images/*/*")

df = pd.DataFrame(files, columns=["file_path"])
df["root"] = df.file_path.str.split("/", expand=True)[0]
df["path"] = df.file_path.str.split("/", expand=True)[1]
df["split"] = df.file_path.str.split("/", expand=True)[2]
df["filename"] = df.file_path.str.split("/", expand=True)[3]
df["extra"] = df.filename.str.split(" ", expand=True)[1]

for i, row in df.query("not extra.isnull()").iterrows():
    os.remove(row.file_path)
```
