```python
import cv2
import matplotlib.pyplot as plt
import random
from glob import glob
from src import walls_detection_cc5k as wdc
```

```python
def plot_images(image1, image2, text1="", text2=""):

    plt.figure(figsize=(15, 15))

    plt.subplot(1, 2, 1)
    plt.title(text1)
    plt.imshow(image1)

    plt.subplot(1, 2, 2)
    plt.title(text2)
    plt.imshow(image2)

    plt.show()


files = glob("data/images/test/*")
files = random.sample(files, 5)

for file in files:
    image = cv2.imread(file)
    image_done = wdc.get_walls(image)

    plot_images(image, image_done, "image original", "image_segmented")
```
